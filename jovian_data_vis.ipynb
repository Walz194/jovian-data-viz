```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.ylabel('Ton per hectre')
plt.xlabel('Year')
```

```python
# For better clarity of what is going on the graph, you can add a legend which will make it more clear as to what each of the
# Lines on the graph represents
plt.plot(years,apples)
plt.plot(years,oranges)

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yield in Oyo')
plt.legend(['APpLe','oRanGe'])
```

```python
# To add markers you can use the marker attr to identify each of the points in the data
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yield in Oyo')
plt.legend(['APpLe','oRanGe'])
```

```python
plt.plot(years,apples,marker='o',ls='--')
plt.plot(years,oranges,marker='s',color='b',ls='--',lw=3,ms=10,alpha=.5)

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yield in Oyo')
plt.legend(['APpLe','oRanGe'])
```

```python
flowers_df = sns.load_dataset('iris')
flowers_df
```

```python
# You might want to find out the relationship between the sepal_length and the sepal_width
# Does the length increase as the width increases?
flowers_df.plot.scatter(x='sepal_length',y='sepal_width')
```

```python
sns.scatterplot(data=flowers_df,x='sepal_length',y='sepal_width',hue='species');
```

```python
# sns is also built on top of matplotlib
plt.figure(figsize=(12,8))
plt.title('Sepal Dimensions')

sns.scatterplot(data=flowers_df,x='sepal_length',y='sepal_width',hue='species',s=100);
plt.legend(loc='upper left')
```

```python
import numpy as np
```

```python
# Histograms are generally used for one column of numeric data. The values in the columns are placed on the xaxis, while
# The y-axis contains the frequency of values within a specific range
plt.title('Distribuition of ')
sns.histplot(data=flowers_df,x='sepal_width')
```

```python
# We can also decide to set custom bins for our histogram
new_interval = np.arange(2,5,0.25)
```

```python
sns.histplot(data=flowers_df,x='sepal_width',bins=new_interval)
```

```python
# We can also plot multiple histograms
# I'd like to split the dataframes into the different species
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
# to make the graphs visible from the ones above, you can reduce the opacity of the other graphs
plt.hist(setosa_df.sepal_width,alpha=.4,label='setosa')
plt.hist(versicolor_df.sepal_width,alpha=.4,label='versicolor')
plt.hist(virginica_df.sepal_width,alpha=.4,label='virginica')
plt.legend();
```

```python
# We can also stack multiple bars on top of each other by passing in an array
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],stacked=True);
```

```python
# Heatmaps: this is used to analyze 2-d data like a matrix or a table
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
# To use this, we are going to want our data to be in a 2d format. Hence we'll use a pivot table to make that possible
flights_df = pd.pivot_table(data=flights,columns='year',index='month',values='passengers')
```

```python
plt.title('Flights over the years')
sns.heatmap(flights_df)
plt.yticks(rotation=0);
```

```python
# We can set an attribut annot=True, to indicate numbers
# We also have cmap; i want lighter color to show a fewer number of passengers and vice-versa
plt.figure(figsize=(12,6))
plt.title('Flights over the years')
sns.heatmap(flights_df,annot=True,cmap='Blues')
plt.yticks(rotation=0);
```

```python
# Plotting multiple plots
fig,axes = plt.subplots(2,3,figsize=(12,9))

axes[0,0].plot(years,oranges,'o--r')
axes[0,0].plot(years,apples,'o--b')
axes[0,0].set_title('Crop yield')
axes[0,0].set_xlabel('Years')
axes[0,0].set_ylabel('Yield(tons per hectare)')
axes[0,0].legend([])

axes[0,1].set_title('Sepal length vs width')
sns.scatterplot(flowers_df.sepal_length,flowers_df.sepal_width,hue=flowers_df.species, ax=axes[0,1])

axes[0,2].set_title('Distribution of Sepal Width')
sns.histplot(data=flowers_df,x='sepal_width',kde=True,ax=axes[0,2])

axes[1,0].set_title('Distribution of Sepal Width 2')
axes[1,0].set_xlabel('Sepal Width')
axes[1,0].set_ylabel('Count')
axes[1,0].hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],stacked=True)
axes[1,0].legend(['Setosa','Vesicolor','Virginica'])

sns.barplot('day','total_bill',hue='sex',data=tips_df,ax=axes[1,1])

sns.heatmap(data=flights_df,ax=axes[1,2])
plt.yticks(rotation=0)

plt.tight_layout(pad=2)
```

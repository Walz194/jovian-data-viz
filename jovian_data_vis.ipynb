```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yield_apples = [0.895,0.91,0.919,0.926,0.929,0.931]
```

```python
# The plot command places the index at the x-axis and plots it against the values
plt.plot(yield_apples)
```

```python
# Adding a semicolon stops the array ([<matplotlib.lines.Line2D at 0x268e9123340>]) from showing
plt.plot(yield_apples);
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.ylabel('Ton per hectre')
plt.xlabel('Year')
```

```python
# For better clarity of what is going on the graph, you can add a legend which will make it more clear as to what each of the
# Lines on the graph represents
plt.plot(years,apples)
plt.plot(years,oranges)

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yield in Oyo')
plt.legend(['APpLe','oRanGe'])
```

```python
# To add markers you can use the marker attr to identify each of the points in the data
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yield in Oyo')
plt.legend(['APpLe','oRanGe'])
```

```python
plt.plot(years,apples,marker='o',ls='--')
plt.plot(years,oranges,marker='s',color='b',ls='--',lw=3,ms=10,alpha=.5)

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yield in Oyo')
plt.legend(['APpLe','oRanGe'])
```

```python
# fmt = '[maeker][line][color]'
plt.plot(years,oranges,'o--r');
```

```python
# Sometime you don't want lines, you can clear the line part of the 'o--r'
plt.plot(years,oranges,'or');
```

```python
flowers_df = sns.load_dataset('iris')
flowers_df
```

```python
# You might want to find out the relationship between the sepal_length and the sepal_width
# Does the length increase as the width increases?
flowers_df.plot.scatter(x='sepal_length',y='sepal_width')
```

```python
sns.scatterplot(data=flowers_df,x='sepal_length',y='sepal_width',hue='species');
```

```python
# sns is also built on top of matplotlib
plt.figure(figsize=(12,8))
plt.title('Sepal Dimensions')

sns.scatterplot(data=flowers_df,x='sepal_length',y='sepal_width',hue='species',s=100);
plt.legend(loc='upper left')
```

```python
# Histograms are generally used for one column of numeric data. The values in the columns are placed on the xaxis, while
# The y-axis contains the frequency of values within a specific range
plt.title('Distribuition of ')
sns.histplot(data=flowers_df,x='sepal_width')
```
